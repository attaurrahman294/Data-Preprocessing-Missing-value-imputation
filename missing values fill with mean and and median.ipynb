# Data cleansing

```python
# get all the informattion about the runing dataset
data.info()
#now we can check that which column has missing values and 
# but this is complex to check one by one we can use multiple methods to check missing values in the below cell
```

```python
# we can also call this function to show  the sum of the all the missing walue
data.isnull().sum()
```

```python
# now find the column numeric column that have missing more than 0 records
missing_num_var = [var for var in data.columns if data[var].isnull().sum() >8]
missing_num_var
```

```python
# we can also plot missing values of graph
plt.figure(figsize=(16,10))
sns.heatmap(data.isnull())
plt.show()
# in the below graph the white color show the highest missing values
```

```python
# now we can also check the column missing values in percent
null_v_per = data.isnull().sum()/data.shape[0]*100
#data.shape[0] = all the records
#data.isnull().sum()  =  sum of all the null in every column
```

```python
null_v_per
```

```python
# Show all the column names which has more then 17% are missing value
drop_column = null_v_per[null_v_per>17].keys()
drop_column
```

```python
# Draf all the column more than 17% missing values
data_c = data.drop(columns=drop_column)
clean_shape = data_c.shape
actual_shape = data.shape
print('Actual shape of hte data is : ', actual_shape)
print('shape of the clean data is :',clean_shape)
```

```python
# to seperate all the column names that has only numeric value
data_rows_num = data_c.select_dtypes(include=['int64','float64']).columns
data_rows_num
```

```python
# to seperate all the column names that has only numeric value
data_df3 = data_c.select_dtypes(include=['int64','float64'])
```

```python
data_df3.head()
```

```python
# we can also plot missing values of graph
plt.figure(figsize=(16,10))
sns.heatmap(data_df3.isnull())
plt.show()
# in the below graph the white color show the highest missing values
```

```python
data_df3.isnull().sum()
```

```python
#show the percentage of missing values
data_df3.isnull().sum()/data_df3.shape[0]*100
```

```python
# now find the column numeric column that have missing more than 0 records
missing_num_var = [var for var in data_df3.columns if data_df3[var].isnull().sum()>0]
missing_num_var
# now we have find that these column has missing values
# we can put mean and mode of these column in the missing cells
```

```python
# draw displot of the data which has null values and 
plt.figure(figsize= (10, 10))
sns.set ()
for i, var in enumerate (missing_num_var):
    plt. subplot (2, 2, i+1)
    sns.distplot (data_df3[var], bins=28)
```

```python
sns.distplot (data_df3['MasVnrArea'], bins=28)
```

```python
# now draw displot for the mssubclass of the clean data
sns.distplot(data_df3['GarageYrBlt'],bins=28)
sns.distplot (data_df3['MasVnrArea'], bins=28)
plt.show()
```

```python
# fill and nan with mean
data_df4_mean = data_df3.fillna(data_df3.mean())
```

```python
data_df4_mean.isnull().sum().sum()
```

```python
# fill all non with mean and compare with non null and null on the graph
plt.figure (figsize= (10, 10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot (2,2, i+1)
    sns.distplot(data_df3[var],bins=20,label="original")
    sns.distplot(data_df4_mean[var],bins=28, label="mean")
    plt.legend ()
```

```python
# fill and nan with mean
data_df5_mean = data_df3.fillna(data_df3.median())
```

```python
# fill all non with mean and compare with non null and null on the graph
plt.figure (figsize= (10, 10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot (2,2, i+1)
    sns.distplot(data_df3[var],bins=20,hist=False,kde_kws={'linewidth':5,'color':'red'},label="original")
    sns.distplot(data_df4_mean[var],bins=28,hist=False, label="mean",kde_kws={'linewidth':4,'color':'blue'})
    sns.distplot(data_df5_mean[var],bins=28,hist=False,kde_kws={'linewidth':4,'color':'green'}, label="median")
    plt.legend ()
```

```python
for i, var in enumerate(missing_num_var):
    plt.figure (figsize= (10, 10))
    plt.subplot (3, 1,1)
    sns. boxplot (data_df3[var])
    plt.subplot (3, 1,2)
    sns.boxplot (data_df4_mean[var ])
    plt.subplot (3,1,3)
    sns.boxplot (data_df5_mean[var] )
```
